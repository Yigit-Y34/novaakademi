```python
import requests

url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd"

response = requests.get(url)

print(response.json())

btc = response.json()["bitcoin"]["usd"]

threshold = 30000

if btc > threshold:
    print("Eşik değerini aşmıştır.")
else:
    print("Eşik değerini aşmamıştır.")
```

```python
import requests

url = "https://restcountries.com/v3.1/name/turkey"

response = requests.get(url)

data = response.json()

country_name = data[0].get('name',{}).get('common','bilgi bulunamadı')
capital = data[0].get('capital',['bilgi bulunamadı'])[0]
region = data[0].get('region','bilgi yok')

subregion = data[0].get('subregion',"bilgi bulunamadı")


print(country_name,capital,region,subregion)
```
