```python
import pandas as pd
```

```python
isimler = ['ali','veli','Mehmet','Fatma']
yaslar = [16, 18,23,28]
sehirler = ['izmir','istanbul','samsun','bursa']
```

```python
data = {'isim': isimler,
        
       'yaş': yaslar,
        
       'şehir': sehirler}

df = pd.DataFrame(data)
```

```python
df
```

```python
print(df['isim'])
```

```python
ilce = ['A','B','C','D']

df['İlçe'] = ilce
```

```python
df
```

```python
yas = df[df['yaş'] > 21]
```

```python
yas
```

```python
sehir = df[df['şehir'] == 'izmir']
sehir
```
