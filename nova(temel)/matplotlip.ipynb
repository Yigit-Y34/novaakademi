```python
categories = ['Elektronik','Giyim','Yiyecek','Mobilya','Kitap']
sales = [3000,1500,2000,3500,2500] #aylık satışlar

plt.bar(categories,sales)
plt.title("Aylık Satış Grafiği")
plt.xlabel("Kategoriler")
plt.ylabel("Satışlar")

for i ,sale in enumerate(sales):
    plt.text(i,sale + 50,str(sale),ha = "center")

plt.show()
```

```python
df = pd.DataFrame({
    'Group': np.random.choice(['A','B','C'],size = 100),
    'Value': np.random.randn(100)
})

ortalama = df.groupby('Group').mean()

plt.plot(ortalama,color = 'r',marker = 'o',linestyle = '--')
plt.title('Grafik')
plt.xlabel('Grup')
plt.ylabel('Değer')
plt.show()

ortalama.plot(kind = 'line',marker = 'o')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Zaman aralığını belirleyin
t = np.arange(0.0, 2.0, 0.01)  # 0'dan 2'ye kadar 0.01 adımlarla

# Sinüs dalga formülü
s = np.sin(2 * np.pi * t)

# Grafiği çizin
plt.plot(t, s)
plt.xlabel('Zaman (s)')
plt.ylabel('Genlik')
plt.title('Sinüs Dalga Grafiği')
plt.grid(True)
plt.show()
```

```python
q1 = np.array([20,35,30,35])
q2 = np.array([25,32,34,28])
q3 = np.array([30,34,37,30])

months = np.array(['Jen','Feb','Mar','Apr'])

df = pd.DataFrame({
    'Months': months,
    'Q1': q1,
    'Q2': q2,
    'Q3': q3
})
df

plt.bar(df['Months'],df['Q1'])
plt.bar(df['Months'],df['Q2'],bottom = df['Q1'])
plt.bar(df['Months'],df['Q3'],bottom = df['Q2'] + df['Q1'])
plt.show()
```

```python
plt.plot(df['Months'],df['Q1'])
plt.plot(df['Months'],df['Q2'])
plt.plot(df['Months'],df['Q3'])
plt.show()
```
